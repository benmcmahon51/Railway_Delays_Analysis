```python
# Setup
import pandas as pd

df0 = pd.read_csv('train_delays_analysis/railway_delays_2020_01_1.csv')
df0.head()
```

```python
# Using a list comprehension to select dataframes from January 1st - January 27th
filenames = ['train_delays_analysis/railway_delays_2020_01_' + str(i) + '.csv' for i in range(1,28)]
filenames
```

```python
# Using a list comprehension to import all dataframes
railway_delays_dfs = [pd.read_csv(filename) for filename in filenames]
railway_delays_dfs[0].info()
```

```python
# Vertically Concatenating all dataframes into a single dataframe
railway_delays_full = pd.concat(railway_delays_dfs)
railway_delays_full.info()
```

```python
# Cleaning and standardizing column names using a for loop
for df in railway_delays_dfs:
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.strip()
    
railway_delays_full = pd.concat(railway_delays_dfs)
railway_delays_full.head()
```

```python
# Cleaning and standardizing column names using a for loop & if/elif statements
for column in railway_delays_full.columns:
    if railway_delays_full[column].dtype == 'object':
        railway_delays_full[column] = railway_delays_full[column].str.upper()
        railway_delays_full[column] = railway_delays_full[column].str.strip()
    elif railway_delays_full[column].dtype == 'float64':
        railway_delays_full[column] = railway_delays_full[column].astype('int64')

railway_delays_full.head()
```

```python
# Investigating if major holidays have an impact on railway delay times
# Counting how many unique days are holidays in the dataset
railway_delays_full[railway_delays_full['major_holiday']==True]['date'].nunique()
```

```python
# Counting how many train rides occured on any of the 4 major holidays
railway_delays_full['major_holiday'].value_counts()
```

```python
# Comparing the average arrival & delay time for trains on major holidays vs non-holidays
railway_delays_full.groupby('major_holiday').agg({'arrival_delay':'mean', 'departure_delay':'mean'}).reset_index()
```

```python
# Analyzing the impact of distance between stations on delay times
railway_delays_full['mileage'].describe()
```

```python
# Analyzing the impact of distance between stations on delay times
# Creating new column for rides with further than average distance
railway_delays_full['long_distance'] = railway_delays_full['mileage'] > 88
railway_delays_full['long_distance'].head(10)
```

```python
# Calculating averiage arrival & departures delays based on distance
railway_delays_full.groupby('long_distance').agg({'arrival_delay':'mean', 'departure_delay':'mean'}).reset_index()
```

```python
# Investigating the impact of weather on delay times
weather = pd.read_csv('train_delays_analysis/weather.csv')
weather.head()
```

```python
# Identifying weather types
weather['weather'].value_counts()
```

```python
# Cleaning and standardizing weather data using for loop & el/elif statements
for column in weather.columns:
    if weather[column].dtype == 'object':
        weather[column] = weather[column].str.upper()
        weather[column] = weather[column].str.strip()
    elif weather[column].dtype == 'float64':
        weather[column] = weather[column].astype('int64')
weather.head()
```

```python
# Joining railway_delays_full and weather using an inner merge
railway_delays_join_weather = pd.merge(railway_delays_full,
                                      weather,
                                      left_on='ride_id',
                                      right_on='ride_id',
                                      how='inner')
railway_delays_join_weather.head()
```

```python
# Calculating average arrival & departure delay times for weather types
railway_delays_join_weather.groupby('weather').agg({'arrival_delay':'mean', 'departure_delay':'mean'})\
.sort_values('arrival_delay', ascending=False).reset_index()
```
